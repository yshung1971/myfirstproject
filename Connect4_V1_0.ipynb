```python
import random

pruning_count = 0
DEBUG = False
# Debug 
def printd(*args, end = '\n'):
    if DEBUG == True:
        print("DBG:",*args, end = end)  



# Create a "row x col = 6 x 7" board
def createBoard():
    board = [[" " for col in range(7)] for row in range(6)]
    return board

def displayBoard(board):
    print(" +---+---+---+---+---+---+---+")
    print(f" | {board[0][0]} | {board[0][1]} | {board[0][2]} | {board[0][3]} | {board[0][4]} | {board[0][5]} | {board[0][6]} |")
    print(" +---+---+---+---+---+---+---+")
    print(f" | {board[1][0]} | {board[1][1]} | {board[1][2]} | {board[1][3]} | {board[1][4]} | {board[1][5]} | {board[1][6]} |")
    print(" +---+---+---+---+---+---+---+")
    print(f" | {board[2][0]} | {board[2][1]} | {board[2][2]} | {board[2][3]} | {board[2][4]} | {board[2][5]} | {board[2][6]} |")
    print(" +---+---+---+---+---+---+---+")
    print(f" | {board[3][0]} | {board[3][1]} | {board[3][2]} | {board[3][3]} | {board[3][4]} | {board[3][5]} | {board[3][6]} |")
    print(" +---+---+---+---+---+---+---+")
    print(f" | {board[4][0]} | {board[4][1]} | {board[4][2]} | {board[4][3]} | {board[4][4]} | {board[4][5]} | {board[4][6]} |")
    print(" +---+---+---+---+---+---+---+")
    print(f" | {board[5][0]} | {board[5][1]} | {board[5][2]} | {board[5][3]} | {board[5][4]} | {board[5][5]} | {board[5][6]} |")
    print(" +---+---+---+---+---+---+---+")
    print("   1   2   3   4   5   6   7")


def checkWin(board):
    # X wins
    # 5,4,3,2,1,0  6x4=24
    for row in range(5,-1,-1):
        # 0,1,2,3: 1,2,3,4: 2,3,4,5: 3,4,5,6
        for col in range(4):
            if board[row][col] == board[row][col + 1] == board[row][col + 2] == board[row][col + 3] == "X":
                return 1000 
    # 6,5,4,3,2,1,0 7x3=21
    for col in range(7):
        # 5,4,3,2: 4,3,2,1: 3,2,1,0
        for row in range(5, 2, -1):
            if board[row][col] == board[row - 1][col] == board[row - 2][col] == board[row - 3][col] == "X":
                return 1000
    # 3x4=12
    for row in range(5,2,-1):
        for col in range(4):
            if board[row][col] == board[row - 1][col + 1] == board[row - 2][col + 2] == board[row - 3][col + 3] == "X":
                return 1000
    # 3x4=12
    for row in range(5,2,-1):
        for col in range(4):
            if board[row][col + 3] == board[row - 1][col + 2] == board[row - 2][col + 1] == board[row -3][col] == "X":
                return 1000

    # O wins
    # 5,4,3,2,1,0  6x4=24
    for row in range(5,-1,-1):
        # 0,1,2,3: 1,2,3,4: 2,3,4,5: 3,4,5,6
        for col in range(4):
            if board[row][col] == board[row][col + 1] == board[row][col + 2] == board[row][col + 3] == "O":
                return -1000
    # 6,5,4,3,2,1,0 7x3=21
    for col in range(7):
        # 5,4,3,2: 4,3,2,1: 3,2,1,0
        for row in range(5, 2, -1):
            if board[row][col] == board[row - 1][col] == board[row - 2][col] == board[row - 3][col] == "O":
                return -1000
    # 3x4=12
    for row in range(5,2,-1):
        for col in range(4):
            if board[row][col] == board[row - 1][col + 1] == board[row - 2][col + 2] == board[row - 3][col + 3] == "O":
                return -1000
    # 3x4=12
    for row in range(5,2,-1):
        for col in range(4):
            if board[row][col + 3] == board[row - 1][col + 2] == board[row - 2][col + 1] == board[row -3][col] == "O":
                return -1000

    # Tie
    count = 0
    for row in range(6):
        for col in range(7):
            if board[row][col] == " ":
                count += 1
    
    if count == 0:
        return 0


def getUserInput(board):
    move = []
    while True:
        user_input = input("Enter a move:(1 to 7) ")
        col = int(user_input.strip())

        if board[0][col-1] != " ":
            print("please enter a valid move")
        else:
            for row in range(5,-1,-1):
                if board[row][col-1] == " ":
                    move = [row, col-1]
                    break
            return move

def findAvailableMoves(board):
    available_moves = []
    for col in range(7):
        for row in range(5, -1, -1):
            if board[row][col] == " ":
                available_moves.append([row, col])
                break

    return available_moves

def getPositionScore(board, turn, row, col):
    score = 0
    printd(f"\nrow = {row}, col = {col}")
    printd("=================\n")

    # Horizontal
    consecutive_square = getConsecutiveSquare(board, 0, row, col)
    score = score + scoreSystem(board,consecutive_square, turn, row, col)

    # Vertical
    consecutive_square = getConsecutiveSquare(board, 1, row, col)
    score = score + scoreSystem(board,consecutive_square, turn, row, col)
    
    # Positive Slope Diagonal
    consecutive_square = getConsecutiveSquare(board, 2, row, col)
    score = score + scoreSystem(board,consecutive_square, turn, row, col)

    # Negative Slope Diagonal
    consecutive_square = getConsecutiveSquare(board, 3, row, col)
    score = score + scoreSystem(board,consecutive_square, turn, row, col)

    # Smarter1 Prediction
    if row > 0:
        printd(f"\nrow = {row}, col = {col}")
        board[row - 1][col] = "O" if turn == "X" else "X"
        printd(f"Step2 Prediction: [row - 1, col] = [{row - 1},{col}]")
        # displayBoard(board)
        consecutive_square = getConsecutiveSquare(board, 4, row - 1, col)
        score = score + step2Score(board,consecutive_square, turn)
        board[row - 1][col] = " "
        printd(f"Step2 Prediction: return to original state: board[row - 1, col] = {board[row - 1][col]}")
        # displayBoard(board)
    return score


def miniMax(board, depth, alpha, beta, isMax):
    if DEBUG:
        global pruning_count
    printd(f"pruning_count = {pruning_count}")

    # Termination State
    # player wins, tie or depth = 0
    score = checkWin(board)
    if score is not None:
        if score == 1000:       # X wins
            printd(f"X wins and score is {score}")
            if DEBUG:
                displayBoard(board)
            return score

        elif score == -1000:    # O wins
            printd(f"O wins and score is {score}")
            if DEBUG:
                displayBoard(board)
            return score

        elif score == 0:        # Tie
            printd(f"Tie and score is {score}")
            if DEBUG:
                displayBoard(board)
            return score
    printd(f"depth = {depth}")
    if depth == 0:
        return 0


    available_moves = findAvailableMoves(board)
    # Max Player
    if isMax == True:
        turn = "X"
        for amoves in available_moves:
            printd(f"turn is {turn}, moves are {available_moves}")
            row = amoves[0]
            col = amoves[1]
            board[row][col] = turn
            score = miniMax(board, depth - 1, alpha, beta, False)  
            board[row][col] = " "
            alpha = max(alpha, score)
            if alpha > beta:
                if DEBUG:
                    pruning_count += 1
                break
        return alpha

    # Min Player
    else:
        turn ="O"
        for amoves in available_moves:
            printd(f"turn is {turn}, moves are {available_moves}")
            row = amoves[0]
            col = amoves[1]
            board[row][col] = turn
            score = miniMax(board, depth - 1, alpha, beta, True)  
            board[row][col] = " "
            beta = min(beta, score)
            if alpha > beta:
                if DEBUG:
                    pruning_count += 1
                break
        return beta        


def getComputerInput(board, turn):
    depth = 5
    move = []
    bs = []

    # Determinate the next round of minimax player
    if turn == "X":
        best_score = -10000
        isMax = False 
    elif turn == "O":
        best_score = 10000
        isMax = True

    available_moves = findAvailableMoves(board)
    for amoves in available_moves:
        row = amoves[0]
        col = amoves[1]
        board[row][col] = turn

        # Get best score and best move
        # default depth is 5. Becareful to set the depth value, if the value is too large,
        # the algorithm will be very very slow, the best value should be 4,5,6
        score = miniMax(board, depth, -1000, 1000, isMax)

        if isMax: # min player 

            # If minimax() returns 0(depth = 0), then use the my design "score system" 
            # to get the best score and the best move.
            # Here is more negative, better
            score = score - getPositionScore(board, turn, row, col)
            if score <= best_score:
                best_score = score

                # Gather all possible moves which score are the same  
                bs.append([row,col,best_score])

        else: # max player
            # If minimax() returns 0(depth = 0), then use the my design "score system" 
            # to get the best score and the best move
            # Here is more positive, better
            score = score + getPositionScore(board, turn, row, col)
            if score >= best_score:
                best_score = score

                # Gather all possible moves which score are the same  
                bs.append([row,col,best_score])                

        # Return the board to its original state
        board[row][col] = " "

        print(f"board[{row}][{col}] score is {score}")
    
    # Sometimes, you will get multiple best_moves. Original design is default use the first one best_move
    # as the "best_move". Here I redesign if we get multiple best moves, just randomly pick up one of them
    # to be the "best_move". It will make game more varied.
    if isMax:
        min_elements = min(bs, key = lambda x: x[2])
        results = [_ for _ in bs if _[2] == min_elements[2]]
    else:
        # pick up the best score one
        max_elements = max(bs, key = lambda x: x[2])

        # find other elements which have the same score as max_element
        results = [_ for _ in bs if _[2] == max_elements[2]]      

    print(f"Candidates: {results}")
    
    # put them into a results pool and randomly pick up one.
    move = random.choice(results)
    print(f"Chosen one: {move}")

    printd(f"getComputerInput(): best_move is {move}, which score is {best_score}")
    return move



def handleMoves(board, turn, game_type):
    # Human vs. Human
    if game_type == 1:
        move = getUserInput(board)
        board[move[0]][move[1]] = turn

    # Computer vs. Human
    elif game_type == 2:
        if turn == "X":
            move = getComputerInput(board, turn)
        else:
            move = getUserInput(board)
        board[move[0]][move[1]] = turn

    # Human vs. Computer
    elif game_type == 3:
        if turn == "X":
            move = getUserInput(board)
        else:
            move = getComputerInput(board, turn)
        board[move[0]][move[1]] = turn

    # Computer vs. Computer
    elif game_type == 4:
        move = getComputerInput(board, turn)
        board[move[0]][move[1]] = turn
    
    return board


def getConsecutiveSquare(board, style, row, col):
    # Get all 4-consecutive-squares indexes combinations
    # style: 0 = Horizontal, 1 = Vertical, 2 = PSD, 3 = NSD
    consecutive_squares = [] # All
    consecutive_square = [] # [row, col] included in each element
    _ = [] # temporary list

    # Horizontal => style = 0
    # Find all horizontal 4-consecutive-squares combinations
    if style == 0 or style == 4:
        printd(f"style is 0 or 4, style = {style}")
        for _row in range(6):
            for _col in range(4):
                for i in range(4):
                    _.append([_row, _col + i]) #  [[,], [,], [,], [,]] 
                consecutive_squares.append(_) # [[[,], [,], [,], [,]], [[,], [,], [,], [,]], ...] 
                printd(f"consecutive_squares: {consecutive_squares} when style = {style}")  
                _ = []

    # Vertical => style = 1
    # Find all vertical 4-consecutive-squares combinations
    if style == 1:
        for _col in range(7):
            for _row in range(5, 2, -1):
                for i in range(4):
                    _.append([_row - i, _col])
                consecutive_squares.append(_)        
                _ = []

    # PSD => style = 2
    # Find all PSD 4-consecutive-squares combinations
    if style == 2 or style == 4:
        printd(f"style is 2 or 4, style = {style}")
        for _row in range(2,-1,-1):
            for _col in range(4):
                for i in range(4):
                    _.append([_row + i , _col + i])
                consecutive_squares.append(_)
                printd(f"consecutive_squares: {consecutive_squares} when style = {style}")  
                _ = []

    # NSD => style = 3
    # Find all NSD 4-consecutive-squares combinations   
    if style == 3 or style == 4:
        printd(f"style is 3 or 4, style = {style}")
        for _col in range(4):
            for _row in range(5, 2, -1):
                for i in range(4):
                    _.append([_row -i , _col + i])
                consecutive_squares.append(_)
                printd(f"consecutive_squares: {consecutive_squares} when style = {style}")  
                _ = []
    
    # Select the elements in consecutive_squares which contains [row,col] 
    # and add them to consecutive_square
    for _ in consecutive_squares:
        if [row,col] in _:
            consecutive_square.append(_)    

    # Return the elements in consecutive_square
    return consecutive_square


def scoreSystem(board,consecutive_square, turn, row, col):
    # Score system:
    # Find all horizontal 4-consecutive-squares combinations (Horizontal, Vertical, Diagonals)
    # Input one of the available moves and check if it's in a 4-consecutive-squares 
    # then give scores based on the following rules:
    # 4 same pieces => 100
    # 3 same pieces + 1 empty square => 3
    # 2 same pieces + 2 empty squares => 1
    # 1 piece + 3 empty squares => 0
    #
    # 4 same opponents pieces => 100
    score = 0
    
    # Contents of the 4-consecutive-squares which available moves included like "X", "O" or " "
    check_contents = [] 
    for _ in consecutive_square:
        for coord in _:
            rr,cc = coord
            check_contents.append(board[rr][cc])

        piece_count = check_contents.count(turn)
        opp_piece_count = check_contents.count("O" if turn == "X" else "X")
        empty_count = check_contents.count(" ")

        printd(check_contents)
        check_contents = [] 
        if opp_piece_count == 4:
            score = score + 250
        if opp_piece_count == 3 and piece_count == 1:
            score = score + 100
        elif opp_piece_count == 2 and piece_count == 1:
            score = score + 10
        elif piece_count == 4:
            score = score + 120
        elif piece_count == 3 and empty_count == 1:
            score = score + 10
        elif piece_count == 2 and empty_count == 2:
            score = score + 3
        elif col == 3: 
            score = score + 5
        elif col == 2 or col == 4:
            score = score + 3
        elif col == 1 or col == 5:
            score = score + 2
        else:
            score = score + 1

    printd(f"Horizontal Score: {score}")
    return score


def step2Score(board, consecutive_square, turn):
    score = 0
    check_contents = []
    printd(f"step2Score: consecutive_square = {consecutive_square}")
    for _ in consecutive_square:
        for coord in _:
            rr,cc = coord
            check_contents.append(board[rr][cc])

        piece_count = check_contents.count(turn)
        opp_piece_count = check_contents.count("O" if turn == "X" else "X")
        empty_count = check_contents.count(" ")

        printd(f"step2Score: check_contents = {check_contents}")
        check_contents = []

        if opp_piece_count == 4:
            score = score - 150
        
        if opp_piece_count == 3 and empty_count == 1:
            score = score - 50

        printd(f"step2Score: score = {score}")
    return score


def playGame(game_type):
    # game_type
    # 1: Human vs. Human
    # 2: Computer vs. Human
    turn = "X"
    board = createBoard()

    # # test 
    # board = [
    #         ['O', ' ', 'X', 'O', 'X', ' ', ' '],
    #         ['X', ' ', 'O', 'X', 'X', ' ', ' '],
    #         ['O', ' ', 'X', 'O', 'O', ' ', ' '],
    #         ['X', ' ', 'O', 'X', 'X', ' ', ' '],
    #         ['O', ' ', 'X', 'O', 'O', ' ', ' '],
    #         ['O', ' ', 'X', 'X', 'O', ' ', ' ']]

    displayBoard(board)

    while True:
        print(f"{turn}'s turn.")  
        # The player who plays X always goes first
        if turn == "X":
            board = handleMoves(board, turn, game_type)
        elif turn == "O":
            board = handleMoves(board, turn, game_type)
            
        displayBoard(board)

        game_state = checkWin(board)
        if game_state is not None:
            if game_state == 1000:
                print("X wins!")
                break
            elif game_state == -1000:
                print("O wins!")
                break
            elif game_state == 0:
                print("Tie!")
                break
        turn = "X" if turn == "O" else "O"

# Usage: playGame(style)
# style =
# 1: Human vs. Human 
# 2: Computer vs. Human
# 3: Human vs. Computer
# 4: Computer vs. Computer

playGame(2)
```
